构建成本计算方程

````python
def best_strategy(tariff):
    cost_japan = base_cost + shipping + base_price * tariff
    cost_mexico = mexico_cost + rail_shipping + base_price * mexico_tariff # USMCA下通常为0
    cost_usa = usa_cost # 高人工，无运费无关税
    return min(cost_japan, cost_mexico, cost_usa)
````

采用人工成本来估计基础成本, 获取到了关税和整车收集

- 采用**加权中位数**计算，确保数据代表性和抗操纵性 

### 1.2 单车海运费计算方法

**数据来源**: 专业汽车运输公司报价和行业报告

| 航线      | 单车海运费估算    | 依据                                          |
| --------- | ----------------- | --------------------------------------------- |
| 东亚→美西 | **$1,000-$1,500** | 40尺集装箱($2,000-3,000)可装2-3辆车，均摊计算 |
| 日本→美国 | **$1,200-$2,000** | West Coast Shipping等专业公司报价             |

**计算公式**:

```
单车海运费 = 集装箱运费 ÷ 每箱装载车辆数
```

**注**: 实际费用受车型、季节、附加费(燃油、港口等)影响，浮动范围约±20%

````python
# 海运运费
# 使用FBX数据计算单车海运费
fbx_rate = 2309  # 2025年9月FBX01指数($/FEU)
vehicles_per_container = 2.5  # 平均每40尺集装箱装载2-3辆轿车
single_vehicle_shipping_cost = fbx_rate / vehicles_per_container

print(f"单车海运费: ${single_vehicle_shipping_cost:.2f}")
````

````python
# 车辆成本
# 基于中国汽车工程学会报告计算成本拆解
car_price = 25000  # 丰田凯美瑞美国售价($)
manufacturing_cost_ratio = 0.63  # 制造成本占比
manufacturing_cost = car_price * manufacturing_cost_ratio

# 成本拆解(取中值)
material_ratio = 0.40  # 材料占比(35%-45%)
labor_ratio = 0.12  # 人工占比(10%-15%)
other_ratio = 0.48  # 其他(研发/管理/利润等)

material_cost = manufacturing_cost * material_ratio
labor_cost = manufacturing_cost * labor_ratio
other_costs = manufacturing_cost * other_ratio

print(f"制造成本: ${manufacturing_cost:.2f} (占售价{manufacturing_cost_ratio*100:.1f}%)")
print(f"材料成本: ${material_cost:.2f} (占制造成本{material_ratio*100:.1f}%)")
print(f"人工成本: ${labor_cost:.2f} (占制造成本{labor_ratio*100:.1f}%)")
print(f"其他成本: ${other_costs:.2f} (占制造成本{other_ratio*100:.1f}%)")
````
